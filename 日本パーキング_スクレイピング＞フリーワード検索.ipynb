```python
import requests
from bs4 import BeautifulSoup

SEARCHQUERY = "品川駅"
request_url = "https://npc-npc.co.jp/parking/search?utf8=%E2%9C%93&search%5Btype%5D=1&word=" + SEARCHQUERY
res = requests.get(request_url)
res.encoding = 'utf-8'
soup = BeautifulSoup(res.text, 'html.parser')
arealist = soup.select('dl.areaList dt a')

i=0

for t in arealist:
    areainfo = soup.select('div.areaListText p')[i:i+4]

    ##############駐車場名##############
    print(t.text)

    ##############空/満情報##############
    print("駐車場ステータス")
    if "駐車場ステータス" in areainfo[0].text:
        print(soup.select('div.areaListText p')[i].select("img")[0]["alt"])  #空/満ステータス
    else:
        print("false")

    ##############最大料金##############
    print("最大料金")
    if "最大料金" in areainfo[0].text:
        print(soup.select('div.areaListText p')[i+1].text)
    else:
        print("None")

    ##############通常料金##############
    print("通常料金")
    if "通常料金" in areainfo[0].text:
        print(soup.select('div.areaListText p')[i+2].text)
    else:
        print("None")

    ##############駐車台数##############
    print("駐車台数")
    if "駐車台数" in areainfo[3].text:
        print(soup.select('div.areaListText p')[i+3].text)
    else:
        print("None")

    print("-----")
    i=i+5
```
