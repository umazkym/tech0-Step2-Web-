```python
import requests
from bs4 import BeautifulSoup
import geocoder
import re
import googlemaps

googleapikey = ""   #GoogleMapAPIkeyを取得
gmaps = googlemaps.Client(key=googleapikey)

SEARCHQUERY = "品川駅"
request_url = "https://npc-npc.co.jp/parking/search?utf8=%E2%9C%93&search%5Btype%5D=1&word=" + SEARCHQUERY
res = requests.get(request_url)
res.encoding = 'utf-8'
soup = BeautifulSoup(res.text, 'html.parser')
arealist = soup.select('dl.areaList dt a')

i=0
j=0
k=0

for t in arealist:
    areainfo = soup.select('div.areaListText p')[i:i+4]

    ##############駐車場名###############
    print(t.text)

    ##############住所###################
    print("住所")
    link = soup.select('div.areaListText p.arrow')[j]
    detail_url="https://npc-npc.co.jp/"+link.a['href']
    res2 = requests.get(detail_url)
    res2.encoding = 'utf-8'
    res2.text
    soup2 = BeautifulSoup(res2.text, 'html.parser')
    arealist2 = soup2.select('table.tableSt tbody tr td')
    for k in range(5):
        if ("都"or"道"or"府"or"県") in arealist2[k].text:
            address=arealist2[k].text

    result = gmaps.geocode(address)
    lat = result[0]["geometry"]["location"]["lat"]
    lng = result[0]["geometry"]["location"]["lng"]
    
    print("緯度")
    print (lat)
    print("緯度")
    print (lng)

    ##############空/満情報##############
    print("駐車場ステータス")
    if "駐車場ステータス" in areainfo[0].text:
        print(soup.select('div.areaListText p')[i].select("img")[0]["alt"])  #空/満ステータス
    else:
        print("false")

    ##############最大料金###############
    print("最大料金")
    if "最大料金" in areainfo[0].text:
        print(soup.select('div.areaListText p')[i+1].text)
    else:
        print("None")

    ##############通常料金###############
    print("通常料金")
    if "通常料金" in areainfo[0].text:
        print(soup.select('div.areaListText p')[i+2].text)
    else:
        print("None")

    ##############駐車台数###############
    print("駐車台数")
    if "駐車台数" in areainfo[3].text:
        print(soup.select('div.areaListText p')[i+3].text)
    else:
        print("None")

    print("-----")
    i=i+5
    j=j+1
```
